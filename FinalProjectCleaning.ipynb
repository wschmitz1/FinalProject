### Preparing for Anova

```python
CrashData_df.dropna(subset=['Weather'], inplace=True)
```

```python
mean_speed_limit = CrashData_df['Speed_Limit'].mean()
CrashData_df['Speed_Limit'].fillna(mean_speed_limit, inplace=True)
```

```python
CrashData_df['Driver_At_Fault'] = CrashData_df['Driver_At_Fault'].map({'Yes': 1, 'No': 0})
```

```python
CrashData_df.dropna(subset=['Driver_At_Fault'], inplace=True)
```
